Loops In Python:
while loop,
for loop

```python
count = 1
while count<=5:
  print("Ashish Loves Coding")
  count += 1
```

```python
i = 1
while i <= 20:
  print(i)
  i += 1
```

```python
i = 20
while i >= 1:
  print(i)
  i -= 1
```

```python
num = [1, 5, 7, 4, 2, 6, 5]
i = 0
while i<len(num):
  print(num[i])
  i += 1
```

```python
num = (44, 25, 96, 36, 15, 75)
x = 36
i = 0
while i<len(num):
  if(num[i] == x):
    print("Founded at",i)
    break
  else:
    print("Finding...")
  i += 1
print("End")
```

```python
i = 1
while i <= 10:
  if(i%2 != 0):
    i+=1
    continue
  print(i)
  i+=1
```
