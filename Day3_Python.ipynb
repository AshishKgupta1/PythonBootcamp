```python
def factorial(num):
    if num < 0:
        return "Negative Number"
    elif num == 0:
        return 1
    else:
        factorial = 1
        for i in range(1, num+1):
            factorial = factorial*i
        return factorial
    
factorial(6)
```

```python
class student:
    def __init__(self, f_name, age, c_grade):
        self.n = f_name
        self.a = age
        self.g = c_grade
        
s1 = student("Ashish", 21, "A")
print(s1.n, s1.a, s1.g)
s2 = student("Ashu", 21, "B")
print(s2.n, s2.a, s2.g)
```
